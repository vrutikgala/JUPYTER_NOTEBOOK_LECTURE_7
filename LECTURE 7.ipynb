```python
report_card_dictionary={"Student_names":["Vanshika","Vivaan","Vrutik","kareena","Mitansh"],
                       "chemistry grades":[23,45,45,67,89],
                       "physics_grades":[23,56,78,98,67],
                       "math_grades":[56,76,87,54,98],
                       "biology_grades":[45,78,98,56,45]}

report_card_dictionary
```

```python
# Setup
import pandas as pd
```

```python
# Making the report card dictionary into a data frame
report_card_df=pd.DataFrame(report_card_dictionary)
report_card_df
```

```python
# First three rows in the data frame
report_card_df.head(3)
```

```python
# random 3 rows in the data frame
report_card_df.sample(3)
```

```python
# Describing all the numeric columns in the report card data frame
report_card_df.describe()
```

# Insights

- What is the average chemistry grade? 53.8
- What is the average biology grade? 64.4
- What is the lowest maths grade? 54
- What is the highest physics grade? 98
- How many scores are present in maths coulmn? 5

```python
# Make a total grade column
report_card_df["total_grades"]=report_card_df["chemistry grades"]+report_card_df["physics_grades"]+report_card_df["math_grades"]+report_card_df["biology_grades"]
report_card_df
```

```python
# Make a column called average grade
report_card_df["average_grade"]=report_card_df["total_grades"]/4
report_card_df
```

```python
report_card_df.describe()
```

```python
report_card_insights_df=report_card_df.describe().transpose()
report_card_insights_df
```

```python
# Saving report card insights as csv file
report_card_insights_df.to_csv("Report Card Insights.csv")
```

```python
report_card_df.loc[report_card_df["total_grades"]>= 140 ,"Status"]="pass"
report_card_df.loc[report_card_df["total_grades"]<140 , "Status"] ="fail"
report_card_df
```

```python
report_card_df.loc[(report_card_df["average_grade"]>= 90),"Character_grade"]="A"
report_card_df.loc[(report_card_df["average_grade"]>= 70) & (report_card_df["average_grade"]<90),"Character_grade"]= "B"
report_card_df.loc[(report_card_df["average_grade"]>= 50) & (report_card_df["average_grade"]<70),"Character_grade"]= "C"
report_card_df.loc[(report_card_df["average_grade"]>= 35) & (report_card_df["average_grade"]<50),"Character_grade"]= "D"
report_card_df.loc[(report_card_df["average_grade"]<35),"Character_grade"]="F"
report_card_df
```

```python
report_card_df.to_csv("Report Card.csv")
```
